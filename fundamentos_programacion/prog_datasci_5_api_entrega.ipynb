Hemos visto que, en realizar una petición a una web API http, recuperamos un objeto que contiene, entre otros, los siguientes atributos: **status.code**, **content** y **headers**. Busca la información sobre los códigos de **status.code** y completa la siguiente tabla sobre los códigos de error http. 

**(1 punto)** <span style="font-family: Courier New; background-color:  #ffcc5c; color: #000000; padding: 3px; ">EG</span>

**Respuesta**

Descripción de los principales códigos de error http:

- 200: The request is correct. Is called "OK". 
- 301: Is one of the redirection related statuses, which indicates that the resource requested has been permanently moved to the URL given by the Location header. Is called "Moved Permanently".
- 400: This response means that the server could not interpret the request given invalid syntax. Is called "Bad Request"
- 401: Authentication is required to get the requested response. This is similar to 403, but in this case, authentication is possible. Is called "Unauthorized"
- 403: The client does not have the necessary permissions for certain content, so the server is refusing to grant an appropriate response.Is called "Forbbiden"
- 404: The server could not find the requested content. This response code is one of the most famous given its high occurrence on the web. Is called "Not Found".
- 505: The version of HTTP used in the request is not supported by the server. Is called "HTTP Version Not Supported"
- 501: The requested method is not supported by the server and cannot be handled. The only methods that servers require support (and therefore should not return this code) are GET and HEAD. Is called "Not Implemented"

Bibliography using is https://developer.mozilla.org/es/docs/Web/HTTP/Status

[congreso.es](http://www.congreso.es/) es la página web del Congreso de los Diputados en España. En ella se guarda una relación de todos los diputados elegidos en cada una de las legislaturas. 

En una de las páginas se puede observar un mapa del hemiciclo, junto con la posición de cada uno de los diputados, su fotografía, su representación territorial y el partido político al que esté adscrito.  Esta url se encuentra en [Hemiciclo](http://www.congreso.es/portal/page/portal/Congreso/Congreso/Diputados/Hemiciclo).

Usad `scrappy` para extraer la siguiente información:

*Nombre*, *Territorio*, *Partido*, *URL Imagen*, en el formato de un diccionario, como por ejemplo:

`{'Nombre': 'Callejas Cano, Juan Antonio ', 'Territorio': 'Diputado por Ciudad Real', 'Partido': 'G.P. Popular en el Congreso', 'url': '/wc/htdocs/web/img/diputados/peq/35_14.jpg'}`

Para Ello: 

- Utilizad el tutorial de scrappy para encontrar un `xpath` que contenga la información requerida
- Extraed la información requerida en forma de diccionario.

**Nota**: si la ejecución del _crawler_ os devuelve un error `ReactorNotRestartable`, reiniciad el núcleo del Notebook (en el menú: `Kernel` - `Restart`)

**(2 puntos)**  <span style="font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px; ">EI</span>

```python
# Library requests and import TextResponse packages
import requests
from scrapy.http import TextResponse

# Library Regular Expression
import re 

# Requests of url

r = requests.get("http://www.congreso.es/portal/page/portal/Congreso/Congreso/Diputados/Hemiciclo")

# Get text of url 
response = TextResponse(r.url, body=r.text, encoding='utf-8')

# Save data of specific path
route_data = response.xpath("//area/@onmouseover").getall()

# Create empty variable
congreso_data = []

# Make a loops to save information 
for i in route_data:
    text = re.findall("'.+?'", i)
    congreso_data.append(
    {"Nombre": re.sub('\(.*?\)', '', text[2]).strip(),
    "Territorio": text[3],
    "Partido": text[-2],
    "Url": text[0]})
```

```python
# Show the values of each person 
congreso_data
```
