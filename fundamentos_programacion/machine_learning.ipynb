```python
# Import numpy and pandas libraries
import numpy as np
import pandas as pd

# Import library dataset
from sklearn import datasets

# Create a diabetes frame
diabetes = datasets.load_diabetes(as_frame=True)
diabetes
```

```python
# 1.1 ejercicio 
df = diabetes["data"]
df["progresion"] = diabetes["target"]
```

```python
# Create new columns with conditional

#Deterioro severo
df["deterioro"] = (diabetes["target"].median() > df["progresion"]).astype("int")

#Deterioro moderado
df["deterioro"] = (diabetes["target"].median() <= df["progresion"]).astype("int")
```

```python
# 1.2 
df.groupby("deterioro").describe()
```

Deterioro 0 es severo y Deterioro 1 es moderado

```python
# 1.3
# Libraries Load
import matplotlib.pyplot as plt
import seaborn as sns

# Multile Plots 
fig, axes = plt.subplots(1,2, figsize=(25, 10))

sns.boxplot(y="age", x="deterioro", data=df, ax=axes[0])
axes[0].set_title("Age")
axes[0].set_xticklabels(("Severo", "Moderado"))

sns.boxplot(y="bmi", x="deterioro", data=df, ax=axes[1])
axes[1].set_title("Bmi")
axes[1].set_xticklabels(("Severo", "Moderado"))

plt.show()
```

En este ejercicio, tendréis que aplicar la regressión lineal múltiple para construir un modelo predictivo de deterioro en pacientes de diabetes según las características del conjunto de datos.

- En primer lugar, tendréis que dividir los datos en conjuntos de entrenamiento (*training*) y evaluación (*test*), con un tamaño de la muestra para test del 33%.

### Model predict deterioro column. Severo is 0 and Moderado 1.

```python
# Import libraris of sklearn: Linear_Model and Model_Selection
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Independent variables
X = df[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']]

# Objective variable
y = df['deterioro']

# Split of sample 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

# Create a model 
reg = LinearRegression()

# Fit model 
reg.fit(X_train, y_train)

# Slope of a line
reg.coef_ 

# When x=0, what is the value of y 
reg.intercept_ 

# Make predictions our X_test data
predictions = reg.predict(X_test)

# R2 is the same that reg.score
r2 = r2_score(y_test, predictions)

print("Intercept: \n", reg.intercept_)
print("Coefficients: \n", reg.coef_)
print("Score: \n", reg.score(X_test, y_test))
print("R2: \n", r2)
```

Realizad un test estadístico [ttest](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html) para responder a la siguiente pregunta: ¿Los hombres diagnosticados con diabetes tienen más tendencia al deterioro que las mujeres?
<br>
Hipótesis Nula: Los hombres diagnosticados con diabetes no tienen una tendencia al deterioro diferente de la que muestran las mujeres. 
<br>
Hipótesis Alternativa: Los hombres diagnosticados con diabetes tienen una tendencia diferente al deterioro que las mujeres.
<br>
Una vez realizado el test, interpreta los valores resultantes y elabora una respuesta. **(1 punto)** <span style = "font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span>

**Nota:** Consulta este [enlace](http://www.conexionismo.com/leer_articulo.php?ref=prueba_t_de_student_para_la_comparacion_de_dos_muestras_independientes-j960497l) si quieres ampliar conocimientos sobre el test estadístico ttest. Para realizar el test estadístico automáticamente podéis utilizar el método [`ttest_ind`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html) de *Scipy.stats*

```python
# Describe of sex
df.groupby("sex").describe()
```

```python
# Import library scipy stats
import scipy.stats as stats
# Male < 0 and Female > 0
t1, p1 = stats.ttest_ind(df.query("sex<=-0.04")["deterioro"], df.query("sex>=0.05")["deterioro"])
```

```python
t1, p1
```

```python
print("Análisis de la evolución de la diabetes por sexo. Estadístico t= " + str(t1) + ", p-val = " + str(p1))
print("Podemos concluir que no rechazamos la hipótesis nula, por tanto no hay diferencia entre hombres y mujeres en la evolución de la diabetes")
```

CHEET: Si observamos un valor p grande, por ejemplo mayor que 0.05 o 0.1, entonces no podemos rechazar la hipótesis nula de puntajes promedio idénticos. Si el valor p es menor que el umbral, por ejemplo, 1%, 5% o 10%, rechazamos la hipótesis nula de promedios iguales.
