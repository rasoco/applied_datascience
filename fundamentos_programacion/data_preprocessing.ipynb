### Ejercicio Opcional

Los valores atípicos son valores que se alejan del resto de variables y, en la mayoría de los casos, son valores que pueden perturbar el resultado final. Por esta razón es necesario identificarlos y, generalmente, eliminarlos del análisis. Utilizad la función `IsolationForest` de la librería [sklearn.ensemble](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html) para determinar los valores atípicos considerando una proporción de valores atípicos en el conjunto de datos de 0,2. Determina cuantos valores atípicos hay en el conjunto de datos derivados del ejercicio 1 y que países deberían de ser eliminados del conjunto de datos final. 


 <span style="font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px; ">EI</span> 

### Create and fit the model

```python
# Import library sklearn and InsolationForest 
from sklearn.ensemble import IsolationForest

# Create model 
model = IsolationForest(n_estimators=100, max_samples="auto", contamination=float(0.2),max_features=1.0)

# Train model 
model.fit(data2[["beer_servings", "spirit_servings","wine_servings", "total_litres_of_pure_alcohol"]])
```

### Using decision_function() and predict()

Decision_function: Average anomaly score of X of the base classifiers.
Predict: Return is an outlier or not, (-1 and 1).

```python
# Average anomaly score of X of the base classifiers
data2["scores"]=model.decision_function(data2[["beer_servings", "spirit_servings","wine_servings", "total_litres_of_pure_alcohol"]])

# Predict if a particular sample is an outlier or not.
data2["anomaly"]=model.predict(data2[["beer_servings", "spirit_servings","wine_servings", "total_litres_of_pure_alcohol"]])
```

### Interpretation about model.predict()
A negative score value and a -1 for the value of anomaly columns indicate the presence of anomaly.  

A value of 1 for the anomaly represents the normal data.


```python
# Show total outliers 
len(data2[data2["anomaly"] == -1])
```

```python
# Show total no outliers
len(data2[data2["anomaly"] == 1])
```

```python
# Save total outliers in variable 
outliers_counter = len(data2[data2["anomaly"] == -1])
```

### Accuracy 

```python
print("Accuracy percentage:", 100*list(data2["anomaly"]).count(-1)/(outliers_counter))
```
